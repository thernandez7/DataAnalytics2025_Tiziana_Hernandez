```r
library(readr)
library(EnvStats)
```

```r
# read data
epi.data <- read_csv("epi_results_2024_pop_gdp.csv")
```

```r
# view dataframe
View(epi.data)
```

```r
# print summary of variables in dataframe
summary(epi.data$LUF.new)
cat("\n")
# print values in variable
epi.data$LUF.new
```

```r
# print summary of variables in dataframe
summary(epi.data$GTI.new)
cat("\n")

# print values in variable
epi.data$GTI.new
```

```r
### Explore Variables LUF and GTI ###

#get cols in own vars
LUF <- epi.data$LUF.new 
GTI <- epi.data$GTI.new 

cat("LUF values: \n")
LUF

cat("\n")

cat("GTI Values: \n")
GTI
```

```r
# boxplot of variables
boxplot(LUF, GTI, names = c("LUF","GTI"))
```

```r
### Histograms ###

# histogram (frequency distribution)
hist(LUF, prob=TRUE)  

# print estimated density curve for variable
lines(density(LUF,na.rm=TRUE,bw=1.)) # or try bw=“SJ”
```

```r
hist(GTI, prob=TRUE)
# lines(density(GTI,na.rm=TRUE,bw=1.)) 

lines(density(GTI,na.rm=TRUE,bw="SJ")) # or try bw=“SJ”
```

```r
# histogram (frequency distribution) over range
hist(LUF, prob=TRUE) 

# range
x_luf <- seq(0., 100., 1)

# Mean and sd 
mu <- mean(LUF, na.rm=TRUE)
sigma <- sd(LUF, na.rm=TRUE)

# generate probability density values for a normal distribution with given mean and sd
d1 <- dnorm(x_luf,mean=mu, sd=sigma,log=FALSE)

# print density values
lines(x_luf,d1)

# print density values - variation
lines(x_luf,.5*d1)
```

```r
# histogram (frequency distribution) over range
hist(GTI, prob=TRUE)

# range
x_gti <- seq(0., 80., 1)

# Mean and sd
mu1 <- mean(GTI, na.rm=TRUE)
sigma1 <- sd(GTI, na.rm=TRUE)


# generate probability density values for a normal distribution with given mean and sd
d2 <- dnorm(x_gti,mean=mu1, sd=sigma1,log=FALSE)

# print density values
lines(x_gti,d2)

# print density values - variation
lines(x_gti,.5*d2)
```

```r
### Empirical Cumulative Distribution Function ###

# plot ecdfs

plot(ecdf(LUF), do.points=FALSE, verticals=TRUE) 
plot(ecdf(GTI), do.points=FALSE, verticals=TRUE) 
```

```r
### Quantile-quantile Plots ###

# print quantile-quantile plot for variable with theoretical normal distribuion
qqnorm(LUF); qqline(LUF)
qqnorm(GTI); qqline(GTI)
```

```r
# print quantile-quantile plot for 2 variables
qqplot(LUF, GTI, xlab = "Q-Q plot for LUF vs GTI") 
qqline(GTI)
```

```r
## Statistical Tests
hist(LUF)
shapiro.test(LUF)
```

```r
hist(GTI)
shapiro.test(GTI) # normality test 
```

```r
#install.packages("nortest") 
library(nortest)
ad.test(LUF) #normality test
ad.test(GTI)
```

```r
ks.test(LUF,GTI) # to test whether two samples came from the same distribution
```

```r
wilcox.test(LUF,GTI) # test that randomly selected values of LUF and GTI from two pops have the same distribution
```

```r
var.test(LUF,GTI)  # FTEST - to determine if variance of samples are significantly different
t.test(LUF,GTI) # Not a good tests- vars dont meet assumptions
```
